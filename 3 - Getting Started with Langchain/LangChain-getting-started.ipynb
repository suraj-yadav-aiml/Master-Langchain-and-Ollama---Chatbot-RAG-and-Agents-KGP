```python
import os
import textwrap

from dotenv import load_dotenv
from langchain_ollama import ChatOllama


load_dotenv()
```

```python
def print_with_wrap(text, max_width=120):
    wrapped_lines = textwrap.wrap(text, width=max_width)

    for line in wrapped_lines:
        if line:
            print(line, end='\n')
```

```python
base_url = "http://localhost:11434"
model = "llama3.2:3b"
# model = "sherlock-homes"
```

```python
llm = ChatOllama(
    base_url=base_url,
    model=model,
    temperature=0.6,
    num_predict=256
)

response = llm.invoke(input="Hi, what can you do ?")

print_with_wrap(response.content)
```

```python
response = ""
for chunk in llm.stream("How to make a Roti ?"):
    response = response + chunk.content
    print(response)
```
